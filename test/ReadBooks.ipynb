```python
# get table index
def getTableIndex(listBooks):

    tableIndex = dict()
    booksInfo = []

    for book in listBooks:
        response_API = requests.get('https://www.gutenberg.org/files/{bookId}/{bookId}-0.txt'.format(bookId=book))
        data = response_API.text
        #### Option 1: Prendre seulement des mots avec carateres de 4 à 10
        words = re.findall(r"[A-Za-z]{4,10}\w+", data)
        occurentCounts = dict()

        for word in words:
            w = word.lower()
            if w in tableIndex:
                if book not in tableIndex[w]:
                    tableIndex[w] += [book]
            else:
                tableIndex[w] = [book]
            if w in occurentCounts:
                occurentCounts[w] += 1
            else:
                occurentCounts[w] = 1
        booksInfo.append({
            "bookId": book,
            "words": occurentCounts,
            "totalWords": len(occurentCounts)
        })

    return tableIndex, booksInfo

```

```python

listBooks = [1,2,3,4,5,6,7,49345,56667]
# listBooks = [56667]
tableIndex, booksInfo = getTableIndex(listBooks)
# [print("data", item['bookId'], item['totalWords']) for item in booksInfo]


def classement_books():
    historyWords =  ["independent","abruptness","cactus", "brother"]
    booksData = []
    for book in booksInfo:
        count = 0
        for w in historyWords:
            if w in book['words']:
                # count number occurs in document
                count += book['words'][w]
                # count if have that word in ducument
                # count += 1
        booksData.append({
            "bookId": book['bookId'],
            "jaccard": (count/book['totalWords'])
        })

    sortedBooks = sorted(booksData, key=lambda d: d['jaccard'], reverse=True) 
    sendBookId = []

    for sb in sortedBooks:
        if sb['jaccard'] > 0:
            sendBookId.append(sb['bookId'])
    return sendBookId

classement_books()
```
